```python
# define a function

def price_label(row):
    if row['prices'] <= 5:
        return 'low-range product'
    elif (row['prices'] > 5) and (row['prices'] <= 15):
        return 'mid-range product'
    elif row['prices'] > 15:
        return 'high-range product'
    else: return 'Not enough data'
```

# If-Statements with For-Loops

```python
# print the frequency of the “orders_day_of_week” column (which day most orders take place)

df_ords_prods_merged['orders_day_of_week'].value_counts(dropna = False)
```

0 means Saturday. This value has the highest frequency, meaning Saturday is the busiest day.
The 4 value has the lowest frequency. A value of 4, refers to Wednesday, meaning Wednesday is the slowest day for Instacart app orders.

```python
# create a result list

result = []

for value in df_ords_prods_merged['orders_day_of_week']:
    if value == 0:
        result.append('Busiest day')
    elif value == 4:
        result.append('Least busy')
    else:
        result.append('Regularly busy')
```

```python
result
```

```python
# create a new column 'busiest day'

df_ords_prods_merged['busiest_day'] = result
```

```python
# check the output

df_ords_prods_merged['busiest_day'].value_counts(dropna = False)
```

## Step 2 create a new column shown two busiest days and two slowest days

```python
# create a result list for busiest days
# two busiest days are the value 0 and 1 in orders_day_of_week column, which mean Saturday and Sunday
# two slowest days are the value 3 and 4 in orders_day_of_week column, which mean Tuesday and Wednesday

result = []

for value in df_ords_prods_merged['orders_day_of_week']:
    if value in (0, 1):
        result.append('busiest days')
    elif value in (3, 4):
        result.append('slowest days')
    else: 
        result.append('regularly busy')
```

## Step 4 Create a new column “busiest_period_of_day”

```python
# check the frequency of the “order_hour_of_day” column

df_ords_prods_merged['order_hour_of_day'].value_counts(dropna = False)
```

```python
df_ords_prods_merged['order_hour_of_day'].describe()
```

Considering the frequecy of 'order_hour_of_day', we can identify the busiest hours of the day into three groups; 
“Most orders” between 10 - 16, 
“Average orders” between 8 - 9 and 17 - 19, 
and “Fewest orders” between 0 - 7 and 20 - 23

```python
# create a result list for busiest period of day

result = []

for value in df_ords_prods_merged['order_hour_of_day']:
    if value in (10,11,12,13,14,15,16):
        result.append('Most orders')
    elif value in (8,9,17,18,19):
        result.append('Average orders')
    else:
        result.append('Fewest orders')
```

```python
# create a new column

df_ords_prods_merged['busiest_period_of_day'] = result
```
